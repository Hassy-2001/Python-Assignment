```python
# use of the function number 1:Void function with one parameter

def print_message(message):
    print(f"Message received: {message}")

# Give Your First Input
given_user_message = input("Enter a message: ")
print_message(given_user_message)



# use of the function number 2: Fruitful function without parameters

import random
def generate_getrandbits():
    return random.getrandbits(10)

# Give Your Second Input
random_number = generate_getrandbits()
print(f"Generated getrandbits of number 10 is: {random_number}")


# use of the function number 3: with at least two parameters and nested conditionals
def calculate_discount(actual_amount, are_you_member):
    # 20% discount for all of those who are our members
    discounted_price = 0.2

    if are_you_member:
        final_amount = actual_amount - (actual_amount * discounted_price)
        print("Membership discount has been applied.Congratulations!!")
    else:
        final_amount = actual_amount
        print("No discount has applied sorry you are not a member!!")

    # Nested conditional depending on final amount of discounted value
    if final_amount > 100:
        # special discount if you purchase over 100$
        return final_amount * 0.95
    else:
        return final_amount

# Give Your Third Input
Price_item = float(input("Enter the price of item :"))
are_you_member = input("Are you a member? (Yes/no) :").lower() == "yes"
discounted_price = calculate_discount(Price_item, are_you_member)
print(f"Final amount which you have to pay after discount is: ${discounted_price:.2f}")


# use of the function number 4: Use while loop within one function
def countdown(number):
    while number > 0:
        print(number)
        number -= 1
    print("Countdown complete!")

# Give Your Fourth Input
count_down_input = int(input("Enter a number for countdown: "))
countdown(count_down_input)


# Anonymous function by using random
generate_anonymous_number = lambda: random.randint(1, 500)

# Give Your Fifth Input
random_number_anonymous = generate_anonymous_number()
print(f"You have Generated random number by using anonymous function: {random_number_anonymous}")

# Collections: List and Dictionary
Chocolate_list = ["Kit Kat", "Toblerone", "Snickers"]
Chocolate_price = {"Kit Kat": 2.00 , "Toblerone": 3.00 , "Snickers": 4.00}

# Advance Method of List and Dictionary
Chocolate_list.append("Twix")
Chocolate_price["Twix"] = 5.00


#Some important methods of List Aand Dictionary

first_string = "Hello,World!"
second_string = "i am doing my assignment!"
uppercase = first_string.upper()
lowercase = first_string.lower()
concat = first_string + second_string
slice = first_string[1:2]


# Displaying collections
print("Choclate Names:", Chocolate_list)
print("Choclate Price:", Chocolate_price)

# some important string methods in python
print("Uppercase Your String:", uppercase)
print("Lowercase Your String:", lowercase)
print("concat Your Two String:", concat)
print("slice Your String:", slice)
```
